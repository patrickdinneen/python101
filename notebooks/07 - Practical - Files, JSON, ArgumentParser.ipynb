# Instructions

1. Write a Python program that you can run from the command line (`python my_code.py`)
2. Your program should accept a single input: a filename to read as input (see [argparse](https://docs.python.org/3/howto/argparse.html))
3. The file will contain newline-separated JSON of simulated events. One JSON document per line. (see [json.loads](https://docs.python.org/3/library/json.html?highlight=json%20loads#json.loads))
4. Your program should read all of the events and output the event types in order of frequency (descending), along with their actual frequency as a percentage of total events.

Other useful references:
[Reading and writing files](https://docs.python.org/3/tutorial/inputoutput.html#reading-and-writing-files)
[Default Dict](https://docs.python.org/3/library/collections.html?highlight=defaultdict#collections.defaultdict)
